```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

aid = pd.read_csv("SDBSAMPLE_A_ID.csv")
pln = pd.read_csv("SDBSAMPLE_P_LN.csv")

aid.drop(['JOIN_SN_TYP'], axis=1, inplace=True)
pln.drop(['JOIN_SN_TYP'], axis=1, inplace=True)
df = pd.merge(aid, pln, on='JOIN_SN', how='right')
df.drop(columns=['IS_KRN', 'IS_ME', 'COM_SN', 'JOIN_SN', 'YM'], inplace=True)
df.rename(columns={'RATE_YN':'label'}, inplace=True)
# df.columns = list(range(df.shape[1]))

train, test = train_test_split(df, test_size=0.2, random_state=42)

print("Train 데이터 개수:", len(train))
print("Test 데이터 개수:", len(test))
```
